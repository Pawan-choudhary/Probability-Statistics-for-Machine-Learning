# ANOVA(Analysis of Variance)

```python
from scipy.stats import f_oneway
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
G1 = norm.rvs(loc=10.0, scale=2.0, size=40)
G2 = norm.rvs(loc=10.3, scale=2.1, size=40)
G3 = norm.rvs(loc=9.8,  scale=2.3, size=50)
```

```python
sns.distplot(G1, hist=False)
sns.distplot(G2, hist=False)
sns.distplot(G3, hist=False)
plt.grid()
```

```python
# H0: Mean are same
# Ha: not same
f_oneway(G1, G2, G3)  # onw way ANOVA test
```

```python
# Changing size
G1 = norm.rvs(loc=10.0, scale=2.0, size=4000)
G2 = norm.rvs(loc=10.3, scale=2.1, size=4000)
G3 = norm.rvs(loc=9.8,  scale=2.3, size=5000)
```

```python
f_oneway(G1, G2, G3)
```

NOTE :  The more the sample size, the more finer differene it can capture

```python
# Changing size
G1 = norm.rvs(loc=10.0, scale=2.0, size=4000)
G2 = norm.rvs(loc=10.0, scale=2.01, size=4000)
G3 = norm.rvs(loc=10.0,  scale=2.03, size=5000)
```

```python
f_oneway(G1, G2, G3)
```

* As sample size increase ANOVA can detact small difference in mu1,mu2,mu3
* The variance in p-vlues depends on sample size.

# Chi-Square Goodness of fit

The chi-square goodness of fit test is a statistical test used to determine whether observed categorical data fits an expected distribution. It is particularly useful when comparing observed data to a theoretical or expected distribution, such as testing whether the observed data follows a specific probability distribution

```python
from scipy.stats import chisquare
from scipy.stats import poisson
import math
```

# Correlation Test

```python
import scipy.stats as stats
import pandas as pf
import numpy as np
import seaborn as sns
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 5, 6, 9, 10]
```

```python
sns.scatterplot(x = x, y=y)
```

```python
# H0: No correlation 
#Perform Pearson correlation test
correlation, p_value = stats.pearsonr(x, y)
```

```python
# Print the correlation coefficient and p-value
print("Correlation coefficient:", correlation)
print("P-value:", p_value)
if p_value<0.05:
  print("Reject H0, x and y are correlated")
else:
  print("Fail to reject H0")  
```
